How about we try another type of plot? This time let's look at a bar plot. First we'll make some data.

```python
# Create a Python dictionary of 3 of your favourite foods with 
# The keys of the dictionary should be the food name and the values their price
```

```python
favorite_food= {"Rice": 20,
                "Spag": 10,
                "Garri": 15
}
```

```python
favorite_food
```

```python
# let check the keys that we have
favorite_food.keys()
```

```python
favorite_food.values()
```

```python
# Create a bar graph where the x-axis is the keys of the dictionary
# and the y-axis is the values of the dictionary


# Add a title, xlabel and ylabel to the plot
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1,
                              figsize=(10, 10))

# let add value to our graphs
ax1.bar(x=list(favorite_food.keys()), height=list(favorite_food.values()))
# let set title to ax1
ax1.set(title="Favorite food bar graph",
        ylabel="Price",
       xlabel="Food")

# Let's make an horizontal bar on our ax2
ax2.barh(y=list(favorite_food.keys()), width=list(favorite_food.values()))

# let customized our graph
ax2.set(title="Favorite Food Horizontal Bar Graph",
       xlabel="Price",
       ylabel="Food");
```
