## Two options for subplots 

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2,
                                            ncols=2,
                                            figsize=(10, 10))

# PLot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax4.hist(np.random.randn(1000));
```

```python
# Subplots options 2
fig, ax= plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10, 10))

# Let plot on diffrent axis
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib object orented (OO) method )

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50= heart_disease[heart_disease["age"] >50]
```

```python
over_50.head()
```

```python
# let check the length
len(over_50)
```

```python
# let;s plot a scatter plot of age aganist cholestora (chol)
# pyplot method
over_50.plot(kind="scatter",
                  x="age",
                  y="chol",
                  c="target",
                  figsize=(10, 6)); # width and height
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax)

ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax= plt.subplots(figsize=(10, 6))

scatter= ax.scatter(x=over_50["age"],
                   y= over_50["chol"],
                   c=over_50["target"])


# Cutomized the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="dashed",
          label="chol mean line");
```

### Subplot of Age, chol and thalach

```python
fig, (ax0, ax1) = plt.subplots(figsize=(10, 15),
                               nrows=2,
                               ncols=1,
                              sharex=True) # shearx allows the two plots to shear the same x label

# Add  data to ax0
scatter1= ax0.scatter(x= over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# customised our graph for ax0
ax0.set(title="Heart Diseses and Cholesterol Levels",
       ylabel="cholesterol")

# let add legened to our axis o
ax0.legend(*scatter1.legend_elements(), title="Target")

ax0.axhline(over_50["chol"].mean(), linestyle="--")

# Now let plot on ax1
scatter2= ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# let add legened to ax1
ax1.legend(*scatter2.legend_elements(), title="Target")

# let customized our plot
ax1.set(title="Heart Diseses and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# let add and horizontal line
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use("seaborn-whitegrid")

# OO method from scratch
fig, ax= plt.subplots(figsize=(10, 6))

scatter= ax.scatter(x=over_50["age"],
                   y= over_50["chol"],
                   c=over_50["target"],
                   cmap="winter") # this change the colour schem


# Cutomized the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--", c="red");
```

```python
# Customizing the y and x axis limitation
fig, (ax0, ax1) = plt.subplots(figsize=(10, 15),
                               nrows=2,
                               ncols=1,
                              sharex=True
                              ) # shearx allows the two plots to shear the same x label

# Add  data to ax0
scatter1= ax0.scatter(x= over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                     cmap="winter")

# customised our graph for ax0
ax0.set(title="Heart Diseses and Cholesterol Levels",
       ylabel="cholesterol")

# set the x axis limint
ax0.set_xlim([50, 80])


# let add legened to our axis o
ax0.legend(*scatter1.legend_elements(), title="Target")

ax0.axhline(over_50["chol"].mean(), linestyle="--", c="red")

# Now let plot on ax1
scatter2= ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# let add legened to ax1
ax1.legend(*scatter2.legend_elements(), title="Target")

# let customized our plot
ax1.set(title="Heart Diseses and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate",)

# set ax1 y_lim
ax1.set_ylim([60, 200])

# let add and horizontal line
ax1.axhline(over_50["thalach"].mean(), linestyle="--", c="red");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
